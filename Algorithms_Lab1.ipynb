```python
import numpy as np
import seaborn as sns
import pandas as pd
from random import randint
import timeit
```

```python
# function returns const
def f_1(v):
    return 1
```

```python
# function returns sum of all elements
def f_2(v):
    return np.sum(v)
```

```python
# function returns product of all elements
def f_3(v):
    return np.prod(v)
```

```python
# function calculates value of polynom in x = 1.5
def f_4_1(v):   
    p = 0
    for i, a in enumerate(v):
        p += (1.5 ** i) * a
    
    return p
```

```python
# function calculates value of polynom in x = 1.5 using Gorner Method
def f_4_2(v):  
    p = v[-1]
    i = len(v) - 2
    while i >= 0:
        p = p * 1.5 + v[i]
        i -= 1
    return p 
```

```python
# bubble sort
def bubbleSort(v):
    length = len(v)
    for i in range(length):
        for j in range(0, length-i-1):
            if v[j] > v[j+1]:
                v[j], v[j+1] = v[j+1], v[j]
```

```python
sns.set(rc={'figure.figsize':(12,8)})
```

```python
ans1 = []
span1 = range(1,2000,1)
for k in span1:
    v = np.random.randint(1, 101, size=k)     
    time = timeit.timeit(lambda: f_1(v), number=5)
    ans1.append(time)
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=span1, y=ans1)
```

```python
def matrix_product(A, B):
    n = len(A)
    C = np.zeros(shape=(n, n))
    for i in range(n):    
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
                
    return C
```

```python
matrix_product_times = []
matrix_x = range(1, 151, 5)
for i in matrix_x:
    matrix_A = np.random.randint(1, 101, size=(i, i))
    matrix_B = np.random.randint(1, 101, size=(i, i))
    product_time = timeit.timeit(lambda: matrix_product(matrix_A, matrix_B), number=5)
    matrix_product_times.append(product_time)
```

```python
sns.lineplot(x=range(1,151,5), y=matrix_product_times, label='experimental')
sns.lineplot(x=matrix_x, y=0.000004 * np.power(matrix_x, 3) + ans7[0], label='theoretical', linewidth=2.5)
```
