```python
import numpy as np
import seaborn as sns
import pandas as pd
from random import randint
import timeit
```

```python
# function returns const
def f_1(v):
    return 1
```

```python
# function returns sum of all elements
def f_2(v):
    return np.sum(v)
```

```python
# function returns product of all elements
def f_3(v):
    return np.prod(v)
```

```python
# function calculates value of polynom in x = 1.5
def f_4_1(v):   
    p = 0
    for i, a in enumerate(v):
        p += (1.5 ** i) * a
    
    return p
```

```python
# function calculates value of polynom in x = 1.5 using Gorner Method
def f_4_2(v):  
    p = v[-1]
    i = len(v) - 2
    while i >= 0:
        p = p * 1.5 + v[i]
        i -= 1
    return p 
```

```python
# bubble sort
def bubbleSort(v):
    length = len(v)
    for i in range(length):
        for j in range(0, length-i-1):
            if v[j] > v[j+1]:
                v[j], v[j+1] = v[j+1], v[j]
```

```python
ans1 = []
span1 = range(1,2000,1)
for k in span1:
    v = np.random.randint(1, 101, size=k)     
    time = timeit.timeit(lambda: f_1(v), number=5)
    ans1.append(time)
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=span1, y=ans1)
```

```python
ans2 = []
span2 = range(1,2000,1)
for k in span2:
    v = np.random.randint(1, 101, size=k)     
    time = timeit.timeit(lambda: f_2(v), number=5)
    ans2.append(time)
    
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=span2, y=ans2)
```

```python
# check complexity
compl = []
length = len(span2)
for i in range(1,length):
    # divide by i to show o(n) complexity
    compl.append(ans2[i] / i)

sns.lineplot(x=span2[1::], y=compl)
```

```python
# only one iteration because bubble sort is too slow
ans5 = []
span5 = range(1,2000,5)
for k in span5:
    v = np.random.randint(1, 101, size=k)     
    time = timeit.timeit(lambda: bubbleSort(v), number=1)
    ans5.append(time)
    
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=span5, y=ans5)
```

```python
# check complexity
compl = []
for i in range(1,len(ans5)):
    # divide by i^2 to show o(n^2) complexity
    compl.append(ans5[i] / (i * i))
sns.lineplot(x=span5[1:len(span5)], y=compl)
```

```python
ans6 = []
span6 = range(1,2000,5)
for k in span6:
    v = np.random.randint(1, 101, size=k)     
    time = timeit.timeit(lambda: np.sort(v, kind='quciksort'), number=5)
    ans6.append(time)
    
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=span6, y=ans6)
```

```python
# check complexity
compl = []
length = len(ans6)
for i in range(2, length):
    # divide by i*log(i) to show o(n*logn) complexity
    compl.append(ans6[i] / (i * np.log(i)))

# I cut first 8 values because of "outliers". Only like that we can see the convergence to some constant 
sns.lineplot(x=span6[10::], y=compl[8::])
```

```python
def matrix_product(A, B):
    n = len(A)
    C = np.zeros(shape=(n, n))
    for i in range(n):    
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
                
    return C
```

```python
matrix_product_times = []
for i in range(1, 151, 5):
    matrix_A = np.random.randint(1, 101, size=(i, i))
    matrix_B = np.random.randint(1, 101, size=(i, i))
    product_time = timeit.timeit(lambda: matrix_product(matrix_A, matrix_B), number=5)
    matrix_product_times.append(product_time)
```

```python
sns.lineplot(x=range(1,151,5), y=matrix_product_times)
```

```python
# check complexity
compl = []
length = len(range(1,151,5))
for i in range(1, length):
    # divide by i**3 to show o(n^3) complexity
    compl.append(matrix_product_times[i] / (i ** 3))

sns.lineplot(x=range(1,151,5)[1::], y=compl)
```
